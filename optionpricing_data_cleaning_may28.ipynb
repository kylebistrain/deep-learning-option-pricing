Feature Engineering

-- T = # trading days / 252 NYSE trading days
-- sigma = rolling average volatility 
-- r risk-free interest rate (DG3MO)

```python
from datetime import datetime
import pandas as pd
import pandas_market_calendars as mcal

# 1. Get the NYSE calendar
nyse = mcal.get_calendar('NYSE')

# 2. Loop over years, collect the valid days and counts
records = []
for year in range(2012, 2025):
    start = f"{year}-01-01"
    end   = f"{year}-12-31"
    
    # valid_days returns a DatetimeIndex of all market‐open timestamps
    days_index = nyse.valid_days(start_date=start, end_date=end)
    
    records.append({
        "year": year,
        "trading_days": days_index,
        "count": len(days_index)
    })

# 3. If you just want a table of counts:
counts_df = pd.DataFrame([
    {"year": rec["year"], "n_trading_days": rec["count"]}
    for rec in records
])

print(counts_df)



```

```python
import pandas as pd
import pandas_market_calendars as mcal

# 0) make sure your dates are pure date (or string 'YYYY-MM-DD') so valid_days() accepts them
combined_data['quotedate']  = pd.to_datetime(combined_data['quotedate']).dt.date
combined_data['expiration'] = pd.to_datetime(combined_data['expiration']).dt.date

# 1) get the NYSE calendar
nyse = mcal.get_calendar('NYSE')

# 2) extract just the unique quote/expiry pairs
pairs = (
    combined_data
    .loc[:, ['quotedate','expiration']]
    .drop_duplicates()
    .reset_index(drop=True)
)

# 3) for each pair, count how many valid days lie between quote and expiry
#    (subtract 1 if you want "days *after* the quote" up to expiry)
def count_td_to_expiry(row):
    days = nyse.valid_days(
        start_date=row['quotedate'].isoformat(),
        end_date  =row['expiration'].isoformat()
    )
    return len(days) - 1

pairs['td_to_expiry'] = pairs.apply(count_td_to_expiry, axis=1)

# 4) merge that back onto your full DataFrame
combined_data = combined_data.merge(
    pairs,
    on=['quotedate','expiration'],
    how='left'
)

# 5) annualize with a 252-day year
combined_data['T'] = combined_data['td_to_expiry'] / 252

# quick check
print("Still missing td_to_expiry:", combined_data['td_to_expiry'].isna().sum())
```

```python
combined_data = combined_data.sort_values('quotedate')

# 1) Collapse to one price per date (take the FIRST, since every row on that date has the same underlying price)
daily = (
    combined_data
    .drop_duplicates('quotedate')
    .set_index('quotedate')
    .loc[:, ['underlying_last']]
)

# 2) Compute log returns on that daily price series
daily['log_ret'] = np.log(daily['underlying_last'] / daily['underlying_last'].shift(1))

# 3) Compute 30‐day rolling vol, annualized
daily['30d_vol'] = (
    daily['log_ret']
      .rolling(window=30, min_periods=30)
      .std(ddof=0)        # population std; use ddof=1 if you prefer sample std
      * np.sqrt(252)
)

# 4) Merge the 30d_vol back into your full dataset on quotedate
combined_data = combined_data.merge(
    daily['30d_vol'].rename('30d_vol'),
    left_on='quotedate',
    right_index=True,
    how='left'
)

# 5) see how many NaNs remain
print("NaNs in 30d_vol:", combined_data['30d_vol'].isna().sum())
```

```python
import yfinance as yf
import pandas as pd
import numpy as np

#Download auto-adjusted SPY and compute daily log-returns
df = yf.download("SPY",
                 start="2010-01-01",
                 end="2024-05-31",
                 auto_adjust=True)
df["logret"] = np.log(df["Close"] / df["Close"].shift(1))

# Compute multiple realized-vol horizons via "log-return-then-annualize"
windows = [10, 22, 44, 63,126,252]   # trading days

for w in windows:
    tau = w / 252.0                            # horizon in years

    # 1) cumulative w-day log-return
    df[f"logret_{w}d"] = df["logret"].rolling(w).sum()

    # 2) std dev of that w-day log-return
    df[f"sigma_{w}d"]  = df[f"logret_{w}d"].rolling(w).std()

    # 3) annualize via σ = σ_τ / sqrt(τ)
    df[f"logvol{w}_annual"] = df[f"sigma_{w}d"] / np.sqrt(tau)

# now you can map directly from your single-level datetime index
for w in windows:
    col = f"logvol{w}_annual"
    combined_data[col] = combined_data["quotedate"].map(df[col])

# sanity check
print(combined_data[[f"logvol{w}_annual" for w in windows]].isnull().sum())
```

```python
combined_data.sort_values('quotedate',inplace=True)
combined_data['quotedate'] = pd.to_datetime(combined_data['quotedate'])


# Perform backward merge-asof to align dividend data to each quote
combined_data_d = pd.merge_asof(
    combined_data,
    div[['EX-DATE', 'rolling_dividend']],
    left_on='quotedate',
    right_on='EX-DATE',
    direction='backward'
)
```

$$
\text{Define the rolling annual dividend from quarterly payouts:}
$$
$$
D_{\text{rolling}} = \sum_{i=1}^{4} D_i
$$
$$
\text{The discrete dividend yield is then:}
$$
$$
y_{\text{discrete}} = \frac{D_{\text{rolling}}}{S}
$$
$$
\text{And the continuous dividend yield becomes:}
$$
$$
q = \ln\left(1 + \frac{D_{\text{rolling}}}{S} \right)
$$


```python


# Example spot price column = 'S'
combined_data_d['discrete_yield'] = combined_data_d['rolling_dividend'] / combined_data_d['underlying_last']
combined_data_d['q'] = np.log(1 + combined_data_d['discrete_yield'])

combined_data_d['is_ex_date'] = (
    combined_data_d['quotedate'] == combined_data_d['EX-DATE']
).astype(int)

combined_data_d.head()
```
